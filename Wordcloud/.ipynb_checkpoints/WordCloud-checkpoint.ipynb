```python
df['Fonte_de_dados'] = df['Fonte_de_dados'].replace('nao descrito', '')
```

```python
import collections
from collections import Counter


lista2=df['Fonte_de_dados'].tolist()
#Removendo as aspas da lista
lista3 = ','.join(lista2).replace("'", "")
#Dividindo cada linha da lista através das vírgulas
palavras2 = lista3.split(',')

#Usando a biblioteca Counter para contar cada elemento (palavra)
contar_tec = Counter(palavras2)#.most_common()
contar_tec
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Import image to np.array
mask = np.array(Image.open('cloud.png'))
stopwords = set(STOPWORDS)
font_path = "Comfortaa-Bold.ttf"
text = " ".join(review for review in df.Fonte_de_dados)
wordcloud = WordCloud(font_path = font_path, width=800, height=400, background_color="white", mask=mask, max_words=119,
                  stopwords=stopwords, repeat=True, regexp=r"\b\w[\w.-]*\b").generate_from_frequencies(contar_tec)
plt.figure(dpi=400)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
wordcloud.to_file("wc.png")
plt.show()
```
