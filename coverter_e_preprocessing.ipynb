#### Renomear colunas

```python
##importanto bibliotecas
import unicodedata
from unicodedata import normalize
import re
```

```python
#Removendo espaços em branco no nome de cada coluna
df.columns = df.columns.str.replace(" ", "_").str.replace("-", "_").str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8')
```

#### Convertendo string para letras minúsculas e removendo acentos

```python
#Definindo função que converte para minúsculas e remove acentos
def pre_processing(text):
    #adaptação do texto
    adp_text = str(text).lower()
    acent_text = normalize('NFKD', adp_text).encode('ASCII','ignore').decode('ASCII')
    return acent_text
```

```python
#Aplicar função para todo o Dataframe
df = df.applymap(pre_processing)
```

```python
#Remover espaços após as vírgulas em todo o Df
df = df.replace(', ', ',', regex=True)
df = df.replace(' ,', ',', regex=True)
#df = df.replace(', ', ',', regex=True)

##Substitui hífen por _ em todo o Df
#df = df.replace('-', '_', regex=True)

##Para colunas específicas
#df.nome_coluna = df.nome_coluna.str.replace("-", "_")
```
